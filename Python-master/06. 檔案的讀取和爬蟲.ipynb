### 標準檔案存取篇

```python
mydic = {"apple":"蘋果", "banana":"香蕉", "water melon":"西瓜"}
```

我們準備把這字典檔存起來。首先我們來看怎麼一一叫出, 是否可找到全部索引關鍵字?

```python
mydic.keys()
```

Python 一個很好用的存檔案式, 是先用 `print` 做, 看來和想像一樣, 就存起來!

```python
for eng in mydic.keys():
    print(f"{eng},{mydic[eng]}")
```

看來不錯, 我們準備照存... 重點就是, 開一個檔案 (比如說 `f`) 準備存檔, 然後把原本 `print` 加上`file=f`。

![print to file](images/print_to_file.png)

我們現在就準備開個檔案 `mydic.csv`, 用來存檔, 所以要給個參數 `w`。

```python
f = open("mydic.csv", 'w')
```

再來就是基本上之前成功的 `print`, 只重加入 `file=f`。

```python
for eng in mydic.keys():
    print(f"{eng},{mydic[eng]}", file=f)
```

再來別忘了關檔。

```python
f.close()
```

#### 讀檔篇

現在我們準備把檔案讀回來。其實 Python 有幾個些許不同的存取方式, 但建議一開始我們先熟悉萬用的一套方法就好了。我們的方式會把整個檔案一次讀回來, 存成一個 list, 這個 list 裡的一個元素就是原來檔案的一行!

```python
f = open('mydic.csv', 'r')
```

```python
my_data = f.readlines()
```

```python
f.close()
```

我們來看看內容。

```python
my_data
```

我們用其中一行來說明等等要做的動作。

```python
line = my_data[0]
```

```python
line
```

首先, 我們需要把右邊的換行符號 `\n` 去掉, 這是用:
    
    line.rstrip('\n')
    
接下來, 要以 `,` 切開字串, 因此再加上 `split(',')`:
    
    line.rstrip('\n').split(',')

```python
line.rstrip('\n').split(',')
```

我們用這個方式, 再做一個字典檔。為了證明沒有作弊, 所以用新的名稱。

```python
newdic = {}

for line in my_data:
    eng, ch = line.rstrip('\n').split(',')
    newdic[eng] = ch
```

```python
newdic
```

成功了!

### Python 系的 pickle 存取

剛剛我們順利的把一個字典檔的資料存入, 又再拿回來。但過程好像有點麻煩, 可不可能我們一下就存入一個字典檔, 再要回來就有個字典可以用呢? 答案是肯定的!

這時要動用 Python 「什麼都可以存」的 `pickle`。

看一眼我們要存的字典。

```python
mydic
```

引入 `pickle` 套件。

```python
import pickle
```

這邊要注意, 存的時候用 `wb` (二進位檔), 讀的時候也是 `rb`。這是因為 `pickle` 會處理比文字還要一般的檔案。

再來就是打開一個 `pickle` 檔, `dump` 下去!

```python
f = open("mydic.pickle", 'wb')
```

```python
pickle.dump(mydic, f)
```

```python
f.close()
```

#### 讀回來也是超容易

```python
f = open("mydic.pickle", 'rb')
```

```python
pickledic = pickle.load(f)
```

```python
f.close()
```

```python
pickledic
```

### 寫個星座運勢程式

現在輸入生日, 可以顯示星座運勢。這基本上是依婉儒老師的想法。

```python
def sign(bday):
    if bday >= 1222:
        return '魔羯'
    elif bday >= 1123:
        return '射手'
    elif bday >= 1023:
        return '天蠍'
    elif bday >= 923:
        return '天秤'
    elif bday >= 823:
        return '處女'
    elif bday >= 722:
        return '獅子'
    elif bday >= 621:
        return '巨蟹'
    elif bday >= 521:
        return '雙子'
    elif bday >= 421:
        return '金牛'
    elif bday >= 321:
        return '牡羊'
    elif bday >= 219:
        return '雙魚'
    elif bday >= 120:
        return '水瓶'
```

```python
def myfortune(月, 日):
    bday = str(月) + str(日).zfill(2)
    mysign = sign(int(bday))
    print("【" + mysign + "座】")
    print(fortune[mysign])
```

```python
myfortune(12, 15)
```
