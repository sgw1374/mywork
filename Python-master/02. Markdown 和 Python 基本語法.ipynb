依然來個標準的 `numpy` + `matplotlib` 套件輸入。

````python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
````

回味失敗的過去...

````python
π = np.pi
````

````python
θ = np.linspace(0, 2*π, 500)
````

````python
r = 3

x = r * np.cos(θ)
y = r * np.sin(θ)
````

````python
plt.plot(x, y)
````

![matplotlib 圖的結構](images/figure_axes.png)

我們要設定一組 axes 的屬性, 就需要知道目前在用的 axes 是哪一組。這要怎麼知道呢? 就是用

    ax = plt.gca()
    
`gca` 的意思是 "Get Current Axes"。

````python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y)
````

果然成功了!

附帶一提, 我們之所以用了 `r=3` 這樣的設定, 是因為如此很容易造出「可變半徑」的極座標系統。比方說, 我們想畫出:

$$r = 1 - \sin(\theta),$$

可以這樣畫出來...

````python
r = 1 - np.sin(θ)

x = r * np.cos(θ)
y = r * np.sin(θ)
````

````python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y)
````

但「藍心」看來怪怪的, `matplotlib` 有可「快速上色法」, 用一個字母就代表一個顏色。

![matplotlib 快速顏色](images/matplotlib_quick_color.png)

````python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y, 'r')
````

![for 迴圈](images/T_for.png)

Python 的 `for` 迴圈基本上就是用 `list` 來做的: 把 `list` 中的元素一個一個拿出來, 執行我們要執行的動作。

````python
X = ['豬', '狗', '牛']
````

````python
for i in X:
    print(i)
````

當然, `for` 迴圈最好的朋友就是 `range`。

````python
for i in range(10):
    print(i)
````

### 問題: 1 加到 100

````python
s = 0

for i in range(1, 101):
    s = s + i
    
print('1 + 2 + ... + 100 =', s)
````

![while 迴圈](images/T_while.png)

        while 條件式:
            條件符合就執行
            
要小心不要讓這個迴圈無窮執行。

````python
n = 1

while n < 10:
    print("*" * n)
    n = n +1
````
