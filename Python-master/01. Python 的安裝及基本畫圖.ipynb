我們現在來畫圖! 這裡我們需要引入套件 (函式庫), 開始的時候我們先不用那麼在意這些指令是什麼意思。大概知道就是要引入兩個套件:

* `numpy`: 做 (數值型的) 數學運算
* `matplotlib`: 畫圖用的標準套件

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

![matplotlib inline](images/matplotlib_inline.png)

我們要學的畫圖方式主要只有兩種:
    
* `plt.plot`: 畫折線圖
* `plt.scatter`: 畫散佈圖

首先是最重要、最常用的 `plt.plot`。

```python
plt.plot([2, -5, 3, 8, -2])
```

我們把 `plt.plot` 引入一個串列 (list) 的數字, 串列都是用中括號。這裡輸入的是:

    [2, -5, 3, 8, -2]

其實是點的 Y 座標。標準的 `plt.plot` 指令是

    plt.plot(X, Y)

其中 X 是點的 X 座標的串列, Y 是點的 Y 座標的串列。

如果我們沒有給 X 座標, X 座標會自動以 0, 1, 2, ... 補上。然後 `plt.plot(X, Y)` 就是把這些點點出來, 再用線段連接起來, 成為折線圖。

```python
# Python 的註解就是像這樣先用井字號
# 這樣 Python 就不會理會我們打什麼
# 接下來我們想把 (0, 2.1), (1.1, 3.4), (2.5, 1), (3.2, 5), (4.8, 7) 這幾個點連起來!
```

```python
X = [0, 1.1, 2.5, 3.2, 4.8]
Y = [2.1, 3.4, 1, 5, 7]
```

```python
plt.plot(X, Y)
```

接著我們開始畫函數! 畫函數基本上就是「描點法」!

![np.linspace](images/linspace.png)

假設我們要畫的第一個函數是

$$f(x) = x^2 + 1$$

X 軸的範圍取 [-10, 10], 取 500 個點。

```python
x = np.linspace(-10, 10, 500)
```

```python
y = x**2 + 1
```

```python
plt.plot(x, y)
```

再來我們畫

$$f(x) = \sin(x)$$

X 軸的範圍取 [0, 6π], 1000 個點。

```python
x = np.linspace(0, 6*np.pi, 1000)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```
