```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
π=np.pi
```

```python
θ=np.linspace(0,2*π,500)
```

```python
r=3
x=r*np.cos(θ)
y=r*np.sin(θ)
```

```python
ax=plt.gca()
ax.set_aspect('equal')
plt.plot(x, y)
```

```python
x=np.linspace(-2*π,2*π,500)
y=np.sin(x)+np.cos(20*x)*0.2
plt.plot(x,y)
```
