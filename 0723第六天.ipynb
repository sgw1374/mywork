# Python 檔案、存檔

```python
mydic = {"apple":"蘋果","banana":"香蕉","water melon":"西瓜"}
```

```python
mydic.keys()
```

```python
for eng in mydic.keys():
    print(f"{eng},{mydic[eng]}")
```

```python
f = open("mydic.csv",'w')
```

```python
for eng in mydic.keys():
    print(f"{eng},{mydic[eng]}",file=f)
```

```python
f.close()
```

### 基本爬蟲方式

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.marieclaire.com.tw/love-sex/astrology"
resp=requests.get(url)
```

```python
print(resp.text)
```

```python
soup = BeautifulSoup(resp.text,'lxml')
```

```python
hot=soup.find_all('div',class_="hot")
```

```python
articles = hot[0].find_all('a')
```

```python
item = articles[0]
```

```python
item.attrs
```

```python
for item in articles:
    if "星星教授安格斯" in item['title']:
        print(item['title'],item['href'])
        print()
```

```python
for item in articles:
    if "星星教授安格斯" in item['title']:
        print(item['title'],item['href'])
        weekly_url = item ['href']
        print()
```

```python
weekly_url
```

```python
web =requests.get(weekly_url)
```

```python
weekly = BeautifulSoup(web.text,'lxml')
```

```python
weekly
```

```python
article = weekly.find_all('div',class_='articleContent')
```

```python
article
```

```python
article = weekly.find_all('div',class_='articleContent')[0]
```
